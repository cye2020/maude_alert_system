# 데이터 다운로드 설명

## 다운로더 로드

```python
import sys
from pathlib import Path

# 상대 경로 사용
project_root = Path.cwd().parent

# 맨 앞에 추가
if str(project_root) in sys.path:
    sys.path.remove(str(project_root))
sys.path.insert(0, str(project_root))

# Python 내장 code 모듈 캐시만 임시 제거
if 'code' in sys.modules:
    del sys.modules['code']

# 이제 import
from code.loading import DataLoader

loader = DataLoader(
    start=2024,
    end=2024,
    output_file='temp.parquet',
    max_workers=4
)
```

## 처음부터 다운로드

```python
# 다운로드 및 로드
loader.process(skip=False)
```

## 스키마 수집까지 끝난 경우

```python
# 다운로드 및 로드
loader.process(skip=True)
```

## 다운로드가 끝난 경우

```python
adapter = 'polars'
polars_kwargs = {
    'use_statistics': True,
    'parallel': 'auto',
    'low_memory': False,
    'rechunk': False,
    'cache': True,
}

df = loader.load(adapter=adapter, **polars_kwargs)
if df is not None:
    print(f"\n데이터프레임 shape: {df.shape}")
    print(df.head())
```
